```python
# Define the path to your CSV file
file_path = ("../Resources/annualweather.csv")  # Modify this path according to your file location

# Read the CSV file into a DataFrame
weather_df = pd.read_csv(file_path)

# Convert the 'datetime' column to pandas datetime format
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])

# Extract month from the 'datetime' column
weather_df['Month'] = weather_df['datetime'].dt.month

# Group by month and calculate monthly statistics
monthly_data = weather_df.groupby('Month').agg({
    'temp': 'mean',   # Average temperature
    'precip': 'sum',  # Total precipitation
    # Add more statistics as needed
}).reset_index()

# Rename month numbers to month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
monthly_data['Month'] = monthly_data['Month'].map(month_names)

# Display the aggregated monthly data
print(monthly_data)
```

```python
import matplotlib.pyplot as plt

# Data
months = ['May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'January', 'February', 'March', 'April']
temperatures = [77.52, 80.55, 82.87, 84.35, 80.98, 75.29, 69.02, 63.86, 61.37, 62.72, 70.60, 71.89]
precipitation = [4.53, 7.10, 6.46, 6.65, 6.45, 2.33, 5.32, 4.59, 1.34, 2.74, 1.51, 0.98]

# Function to create a chart
def create_chart(x, y, title, xlabel, ylabel, color, marker=None, linestyle=None):
    plt.figure(figsize=(10, 6))
    if marker and linestyle:
        plt.plot(x, y, marker=marker, color=color, linestyle=linestyle)
    else:
        plt.bar(x, y, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Create a bar chart for temperatures
create_chart(months, temperatures, 'Average Monthly Temperature', 'Month', 'Temperature (F)', 'skyblue')

# Create a line plot for precipitation
create_chart(months, precipitation, 'Monthly Precipitation', 'Month', 'Precipitation (inches)', 'orange', marker='o', linestyle='-')
```

```python

import matplotlib.pyplot as plt
import pandas as pd

# Your data frame
google_trend_df = pd.read_csv("../Resources/GoogleTrendsTimeline.csv")

# Convert 'Datetime' column to datetime type
google_trend_df['Datetime'] = pd.to_datetime(google_trend_df['Datetime'])

# Extract month and year from datetime
google_trend_df['Month-Year'] = google_trend_df['Datetime'].dt.to_period('M')

# Group by month-year and calculate the mean of 'Florida Mall' visits
monthly_data = google_trend_df.groupby('Month-Year')['Disney Parks'].mean()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
monthly_data.plot(kind='bar', color='skyblue')
plt.title('Average Monthly Visits to Disney Parks')
plt.xlabel('Month-Year')
plt.ylabel('Average Visits')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
